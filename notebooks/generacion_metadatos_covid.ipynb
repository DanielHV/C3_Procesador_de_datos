```python
import pandas as pd
from unidecode import unidecode
import re
```

```python
df = pd.read_csv("../data/covid19/raw/COVID19MEXICO.csv")
dicc = pd.read_excel("../data/covid19/raw/240708 Descriptores_.xlsx")
catalogos = pd.read_excel("../data/covid19/raw/240708 Catalogos.xlsx", sheet_name=None)
```

---

Se cambian los nombres de la columna `FORMATO O FUENTE` en el diccionario para eliminar puntuaciones y espacios inconsistentes

```python
dicc["FORMATO O FUENTE"] = dicc["FORMATO O FUENTE"].str.replace(":", "", regex=False).str.strip()
dicc["FORMATO O FUENTE"] = dicc["FORMATO O FUENTE"].apply(unidecode)
dicc["FORMATO O FUENTE"] = dicc["FORMATO O FUENTE"].str.replace(r'\s*_\s*', '_', regex=True)
```

```python
dicc["FORMATO O FUENTE"]
```

Se cambian los nombres de las llaves de los catálogos para eliminar puntuaciones, espacios inconsistentes y convertir a mayúsculas

```python
catalogos = {unidecode(llave).upper(): valor for llave, valor in catalogos.items()} # conversion a mayusculas
catalogos = { # conversion de cadena 'CAT' a 'CATALOGO'
    re.sub(r'^CAT ', 'CATALOGO ', llave): valor
    for llave, valor in catalogos.items()
}
```

```python
catalogos
```

Se agrega al diccionario una columna `POSIBLES VALORES` que incluye una lista de valores según el catálogo que le corresponde, si no se indica un catálogo válido, será una lista vacía

```python
def obtener_posibles_valores(catalogo):
    try:
        df_catalogo = catalogos[catalogo]
        if "CLASIFICACIÓN" in df_catalogo.columns:
            return list(df_catalogo["CLASIFICACIÓN"])
        elif "DESCRIPCIÓN" in df_catalogo.columns:
            return list(df_catalogo["DESCRIPCIÓN"])
        else:
            return []
    except KeyError:
        return []
    
def obtener_posibles_valores_alias(catalogo):
    try:
        return list(catalogos[catalogo]["CLAVE"])
    except KeyError:
        return []
```

```python

dicc["POSIBLES VALORES"] = dicc["FORMATO O FUENTE"].apply(obtener_posibles_valores)
dicc["POSIBLES VALORES ALIAS"] = dicc["FORMATO O FUENTE"].apply(obtener_posibles_valores_alias)
```

```python
dicc
```

```python
dicc.to_csv("../data/covid19/raw/240708 Descriptores_rework.csv", index=False)
```

En el archivo de datos, se modifican los valores de las columnas correspondientes a entidades para respetar el formato de cadenas

```python
df["ENTIDAD_UM"] = df["ENTIDAD_UM"].astype(str).str.zfill(2)
df["ENTIDAD_NAC"] = df["ENTIDAD_NAC"].astype(str).str.zfill(2)
df["ENTIDAD_RES"] = df["ENTIDAD_RES"].astype(str).str.zfill(2)
df["MUNICIPIO_RES"] = df["MUNICIPIO_RES"].astype(str).str.zfill(3)
```
