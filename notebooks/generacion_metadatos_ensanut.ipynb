# Generación de metadatos ENSANUT

Se genera el archivo de metadatos necesario para el programa de preprocesamiento.

```python
import pandas as pd
import ast
```

```python
metadatos_ensanut = pd.read_csv('../data/ensanut/raw/21ensanut_a20_hias.csv', dtype=str)
```

```python
metadatos_ensanut.head()
```

```python
metadatos_ensanut['values_dict'] = metadatos_ensanut['values'].apply(lambda x: ast.literal_eval(x))
posibles_valores = pd.json_normalize(metadatos_ensanut['values_dict'].tolist(), max_level=0)
posibles_valores
```

```python
# se crea una nueva columna de posibles valores, y otra de posibles valores pero con sus respectivos alias

metadatos_ensanut['posibles_valores_alias'] = metadatos_ensanut['values_dict'].apply(lambda x: list(x.get('options', {}).keys()))
metadatos_ensanut['posibles_valores'] = metadatos_ensanut['values_dict'].apply(lambda x: list(x.get('options', {}).values()))
metadatos_ensanut['posibles_valores_alias'] = metadatos_ensanut['posibles_valores_alias'].astype(str)
metadatos_ensanut['posibles_valores'] = metadatos_ensanut['posibles_valores'].astype(str)
```

```python
metadatos_ensanut['options'] = posibles_valores['options']

# se cambia el tipo de la variable municipio a options
metadatos_ensanut.loc[metadatos_ensanut['var'] == 'municipio', 'var_type'] = 'options'

metadatos_ensanut.head()
```

```python
metadatos_ensanut['dtype'] = 'str'
metadatos_ensanut.head()
```
